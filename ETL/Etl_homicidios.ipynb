Observamos 608 datos de tipo datatime.time, 85 de tipo string y 3 de tipo datetime.datetime. Se decide dejar todos los datos de tipo datetime.time ya que somn los que mayor cantidad hay

Cambiamos los datos a tipo datetime.time, pero nos sigue faltando uno que no se pudo convertir

Obsrvamos que el hay un dato en le columna Hora que aparece como None, y en la columnas Hora_entero como SD, por lo tanto este puede ser el dato que nos esta faltanto en ambas columnas. Para la columna Hora se decide imputar mediante medidas de tendencia central, especificamente por la moda

```python
# Calculamos la moda
hora_moda = homicidios_hechos["Hora"].mode().iloc[0]
```

```python
# Reemplazamos
homicidios_hechos["Hora"].fillna(hora_moda, inplace=True)
```

```python
# Verificamos
homicidios_hechos["Hora"].apply(type).value_counts()
```

En este proceso calculamos la hora mas comun o la moda de la columna para imputar el valor que aun teniamos como nulo
